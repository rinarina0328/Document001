```
h = ts[1] - ts[0]
```

```
h
```

```
ws = np.arange(0,10,0.01)
Re = np.empty(ws.size)
Im = np.empty(ws.size)
Fs = np.empty(ws.size)
```

```
n = 0
for w in ws:
  Re[n]=xs[n]*np.cos(ws[n]*ts[n])*h
  Im[n]=xs[n]*np.sin(ws[n]*ts[n])*h
  Fs[n]=np.sqrt(Re[n]*Re[n]+Im[n]*Im[n])
  n += 1
```

```
Fs
```

```
np.savetxt('out.txt', np.c_[ws, Fs])
```
